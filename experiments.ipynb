```python
#Importing Libraries
import pandas as pd #for reading the file
from sklearn.model_selection import train_test_split # forsplittingthe data
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder ##for standarization
import pickle #used for reusingthe pickle files
```

```python
#Load the dataset
data=pd.read_csv("Churn_Modelling.csv")
data.head(5)
```

```python
## Preprocess the data

#Droping the irrelevent data i.e 3 columns from starting
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1) #default axis is 0 i.e rows
```

```python
## Encoding Categorical variable i.e geography and gender
LabelEncoder_gender=LabelEncoder() # it will assign numbers from 0 to n
data['Gender']=LabelEncoder_gender.fit_transform(data[['Gender']])
```

```python
#One hot encoder for geography
onehot_geo_encoder=OneHotEncoder()# One hot encoder will make separate columns 
geo_encoder=onehot_geo_encoder.fit_transform(data[['Geography']]).toarray()# It will convert into array
geo_encoded_df=pd.DataFrame(geo_encoder,columns=onehot_geo_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
geo_encoded_df
```

```python
#adding the columns and droping geography
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
data
```

```python
###Save the Encoders and scalar
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(LabelEncoder_gender,file)
with open('onehot_geo_encoder.pkl','wb') as file:
    pickle.dump(onehot_geo_encoder,file)
```

```python
# Divide the data into independent and dependent features
x=data.drop('Exited',axis=1)
y=data['Exited']

##Split the data in training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#Scaling the features
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

### ANN Implementation

1]Sequential network
2] Using Dense for creating hidden layers
3]Applying Activation function in every layer i.e sigmoid,relu,leaky relu
4]Applying Optimizer -useful in backpropogation when updating weights
5]Loss Function -using gradient descent
6]Applying metrics > Accuracy ,mean sqaure error ,f1 score
7]Training information will be stored in logs and folder using tensorboard


```python
# importing Libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential #importing model
from tensorflow.keras.layers import Dense  # for hinden layers
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard 
import datetime
```

```python
## Build ANN Sequential Model 
# Inside Model inputs will be dense layers
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), ## 1st Hidden layer connected with input layer ,input size is only for 1st hidden layer
    Dense(32,activation='relu'),##Dense(neurons_in_hidden_layer, activation=fucntion_name)
    Dense(1,activation='sigmoid')

])
```

```python
model.summary()
```

```python
#another way of using optimers
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
```

```python
## Compile the Model
model.compile(optimizer=opt,loss=loss,metrics=["accuracy"])
```

```python
## Set up the tensorboard to write logs while training 
log_dir="log/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# Setup early Stopping to stop when the model is trained before e.g 100 epocks and trained at 60 epoches
early_stopping_callback=EarlyStopping(monitor='val_losss',patience=10,mode='min',restore_best_weights=True)
```

```python
## Training the Model
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
#saving the model
model.save('model.h5') #h5 file is compatible for keras
```
